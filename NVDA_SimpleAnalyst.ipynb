```python
import pandas as pd
```

```python
dt = pd.read_csv("NVDA.csv")
summary_simple = dt.describe()
summary_simple
```

```python
Most_High = summary_simple["High"].max()
Most_Low = summary_simple["Low"].min()
Range = Most_High - Most_Low
```

```python
Close_High = summary_simple["Close"].max()
Close_Low = summary_simple["Close"].min()
Close_Range = Close_High - Close_Low
```

```python
X = Most_High - Close_High
y = Most_Low - Close_Low
z = Range - Close_Range
```

```python
def Summary():
    result = f'''
    ---Summary---
    Close High:{Close_High}  Close Low:{Close_Low}   Close Range:{Close_Range}
    Most High:{Most_High}   Most Low:{Most_Low} Range_Most:{Range}
    Subtract_High:{X} Subtract_Low:{y}  Subtract_Range:{z}
    '''
    print(result) 
```

```python
Summary()
```
